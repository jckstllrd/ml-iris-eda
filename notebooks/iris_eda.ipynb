## Step 2: Data Preparation

```python
df = df.drop(['Id'], axis=1)
```

```python
df.loc[df.duplicated(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])]
```

# Step 3: Feature Understanding

(Univariate analysis)

- Plotting Feature Distributions
    - Histogram
    - KDE
    - Boxplot

```python
ax = df['SepalLengthCm'].value_counts() \
    .head(10) \
        .plot(kind='bar', title="Length of Iris Sepal (Cm)")
        
ax.set_xlabel('Sepal Length (Cm)')
ax.set_ylabel('Count')
```

```python
ax = df['PetalWidthCm'].plot(kind='hist', bins=20, title='Petal Width (Cm)')

ax.set_xlabel('Width of Petal (Cm)')
```

```python
ax1 = df['SepalLengthCm'].plot(kind='kde', title='Sepal Length (Cm)')
ax2 = df['SepalWidthCm'].plot(kind='kde', title='Sepal Width (Cm)')
ax3 = df['PetalLengthCm'].plot(kind='kde', title='Petal Length (Cm)')
ax4 = df['PetalWidthCm'].plot(kind='kde', title='Petal Width (Cm)')



```

```python
sns.pairplot(df, vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

plt.show()
```

```python
df_corr = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

## Current Thoughts - Write up

This dataset contains 150 entries split equally between the three flower species.

4 numeric features + 1 categorical feature (Species)



### 2. Class Balance

The classes are balanced, 3 duplicate entries were removed, 2 in Setosa, 1 from Virginica.

Leaving -
    - Setosa: 48 entries
    - Versicolor: 50 entries
    - Virginica: 49 entries


### 3. Features

- For the single axis plots, Sepal Width and Sepal Length have some major overlap across each of the flower varieties, these features may not be very strong for classifying the flowers.
- Petal Length starts to show some differentiation between Versicolor and Virginica, it may be a more promising feature in combination with the Petal Width which also shows some disparity between the similar iris variants.

- Based on the plots above, all of the three flowers have a similar distribution of Sepal Length to Sepal Width - this will not be too helpful in classifiying them.
- When comparing the Sepal to the Petal, the differences become much clearer to distinguish between the Setosa Iris against the Versicolor and Virginica variants.
- The Versicolor and Virginica variants of the iris have similar distributions across all of the feature comparisons.
    - The comparison with the biggest visual difference between the Setosa and the two similar variants is the Petal Length and Petal Width Comparison
    - The Versicolor and Virginica might be difficult to differentiate.

Summary:
- Petal Length: Strong
- Petal Width: Strong
- Sepal Length: Weak
- Sepal Width: Weak


### 4. Relationships

Petal Length and Petal Width are strongly correlated (0.96) and so are mostly redundant.

Sepal Length is fairly strongly correlated to both Petal Length (0.87) and Petal Width (0.82).

Sepal Width generally shows limited correlation against all other features so it is relatively independent.
